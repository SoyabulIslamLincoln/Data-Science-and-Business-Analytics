### Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
  
```

### Getting Data

```python
data=pd.read_csv('Iris.csv', index_col='Id')
data.head(10)
```

### Setting Train and Test Data for model

```python
x= data.iloc[:, 0:4]
y=data.iloc[:, -1]
print(x.head())
print(y.head())
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=.20, random_state=24)
```

## Decision Tree Classification

```python
decision_tree = DecisionTreeClassifier()

decision_tree.fit(x_train, y_train)
```

### Visualizing Decision Tree

### 

```python
figure, ax= plt.subplots( figsize=(20,20), dpi= 500)
tree.plot_tree(decision_tree, feature_names=x.columns,class_names=['Iris-setosa', 'Iris-versicolor','Iris-virginica'],
                filled=True, rounded=True,)
figure.show()
```

### Predicted Value

```python
pred=decision_tree.predict(x_test)

print(f'Test precictions: {pred}')
```

### Prediction Accuracy

```python
accuracy= accuracy_score(y_test,pred)
print(f'Prediction Accuracy: {accuracy}')
```
